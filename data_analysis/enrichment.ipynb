## enrich the dataframe using a sample from that dataframe

- create a sample with 200 new observations with random_state = 21

the sample should not have new combinations of the car number,
make
and model, so the whole dataset will be consistent in these terms
there are no restrictions on the refund and fines, you can take
any value
from these columns at random and use it towards any car number


- concatenate the sample with the initial dataframe to a new dataframe con-
cat_rows

```python
sample = df.sample(n=200, random_state=21).reset_index(drop=True)
sample
```

```python
sample_for_Refund = df.sample(n=200, random_state=42).reset_index(drop=True)
sample_for_Refund['Refund']
```

```python
sample_for_Fines = df.sample(n=200, random_state=30).reset_index(drop=True)
sample_for_Fines['Fines']
```

```python
sample['Fines'] = sample_for_Fines['Fines']
sample['Refund'] = sample_for_Refund['Refund']
sample
```

```python
concat_rows = pd.concat([df, sample])
concat_rows
```

```python
surname = pd.read_json('data/surname.json', orient='columns')
surname.columns = surname.iloc[0]
surname = surname.drop(0).reset_index(drop=True)
surname
```

```python
owner_car = concat_rows.drop_duplicates('CarNumber')['CarNumber']
owner_car = owner_car.reset_index(drop=True)
owner_car
```

```python
owner_name = surname['NAME'].sample(n=531, random_state=21, replace=True, ignore_index=True)
owner_name
```

```python
owners = pd.DataFrame(data=owner_car)
owners['SURNAME'] = owner_name
owners
```

```python
data1 = pd.Series(['PHAHA777RUS', 'AHAHA777RUS', 'OLOLO777RUS','NANANA777RUS', 'BINGO777RUS'], name='CarNumber')
data2 = pd.Series([np.random.choice(fines['Refund']) for i in range(5)], name= 'Refund')
data3 = pd.Series([np.random.choice(fines['Fines']) for i in range(5)], name= 'Fines')
data4 = pd.Series([np.random.choice(fines['Make']) for i in range(5)], name= 'Make')
data5 = pd.Series([np.random.choice(fines['Model']) for i in range(5)], name= 'Model')
data6 = pd.Series([np.random.choice(fines['Year']) for i in range(5)], name= 'Year')

df2 = pd.DataFrame([data1, data2, data3, data4, data5, data6])
fines = pd.concat([fines, df2.T]).reset_index(drop=True) # transpose
fines
```

```python
new_owners = owners.drop(owners.tail(20).index)
data1 = pd.Series(['S21MSK005RUS', 'S21NSK006RUS', 'S21KZN007RUS'], name='CarNumber')
data2 = pd.Series([np.random.choice(owners['SURNAME']) for i in range(3)], name= 'SURNAME')
add3 = pd.DataFrame([data1, data2])
new_owners = pd.concat([new_owners, add3.T]).reset_index(drop=True)
new_owners.tail
```

```python
common_cars1 = pd.merge(fines, new_owners, on='CarNumber', how='inner')
common_cars1
```

```python
common_cars2 = pd.merge(fines, new_owners, on='CarNumber', how='outer')
common_cars2
```

```python
common_cars3 = pd.merge(fines, new_owners, on='CarNumber', how='left')
common_cars3
```

```python
common_cars4 = pd.merge(fines, new_owners, on='CarNumber', how='right')
common_cars4
```

```python
pd.pivot_table(
    fines,
    values='Fines',
    index=['Make', 'Model'], 
    columns='Year',
    aggfunc='sum',
)
```

```python
fines.to_csv('data/fines.csv', index=False)
new_owners.to_csv('data/owners.csv', index=False)
```
